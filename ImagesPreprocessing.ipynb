Image Processing with Opencv(Computer vision)



CV2 will always read image as bgr

```python
import cv2
```

```python
path= "/content/CuteCat.jpeg"
imag1= cv2.imread(path)
imag1
```

```python
path= "/content/Alphabet Example.jpg"
imag2= cv2.imread(path)
imag2
```

```python
import matplotlib.pyplot as plt
plt.imshow(imag1)
```

Convert BGR to RGB

```python
img2= cv2.cvtColor(imag2, cv2.COLOR_BGR2RGB)
plt.imshow(imag2)
```

```python
img1= cv2.cvtColor(imag1, cv2.COLOR_BGR2RGB)
plt.imshow(imag1)
```
