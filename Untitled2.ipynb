Step 4 : Perform preprocessing on X

```python
X.dtypes
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='median')),('scaler',StandardScaler())]).set_output(transform='pandas')
```

```python
  X_pre = num_pipe.fit_transform(X)
  X_pre.head()
```

Step : 5 Train test split

```python
from sklearn.model_selection import train_test_split
X_train,Y_train,X_test,Y_test=train_test_split(X_pre,Y,test_size=0.2,random_state=42)
```

Building Neuron Network
Dropout layer - Randomly drop some % of neuron links

```python
from keras.models import Sequential
from keras.layers import Dense,Input,Dropout
```

```python
# Initialize the model
model=Sequential()
```

```python
# Initialize the model
model = Sequential()

# Add the input layer and the first hidden layer

model.add(Input(shape=(X_train.shape[1],)))

# Hidden layer 1
model.add(Dense(64 , activation='relu'))
# add Dropout layer 1
model.add(Dropout(0.4))

# Hidden layer 2
model.add(Dense(32, activation='relu'))
# add Dropout layer 2
model.add(Dropout(0.3))

# Add Outer layer
model.add(Dense(1, activation='sigmoid'))

model.summary()
```
